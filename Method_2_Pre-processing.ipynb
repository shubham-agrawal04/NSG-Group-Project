```python
import numpy as np, gc, os

def normalize_per_sample(X):
    X_min = X.min(axis=(1,2,3), keepdims=True)
    X_max = X.max(axis=(1,2,3), keepdims=True)
    return (X - X_min) / (X_max - X_min + 1e-8)

def add_coords_channel(X):
    b, c, h, w = X.shape
    yy, xx = np.meshgrid(np.linspace(0, 1, h), np.linspace(0, 1, w), indexing='ij')
    yy = np.repeat(yy[None, None, :, :], b, axis=0)
    xx = np.repeat(xx[None, None, :, :], b, axis=0)
    return np.concatenate([X, yy, xx], axis=1)

def preprocess_and_save(path, outdir):
    os.makedirs(outdir, exist_ok=True)
    for i in range(1, 6):
        x = np.load(f"/kaggle/input/openfwi-velocity-data/{path}/data/data{i}.npy")
        y = np.load(f"/kaggle/input/openfwi-velocity-data/{path}/model/model{i}.npy")

        x = normalize_per_sample(x)
        x = add_coords_channel(x)

        np.save(f"{outdir}/X{i}.npy", x)
        np.save(f"{outdir}/Y{i}.npy", y)

        del x, y
        gc.collect()
    print(f"Saved preprocessed chunks to {outdir}")

# preprocess_and_save("FlatVel_A", "FlatVel_A_preprocessed")
preprocess_and_save("FlatVel_B", "FlatVel_B_preprocessed")
# preprocess_and_save("CurveVel_A", "CurveVel_A_preprocessed")
# preprocess_and_save("CurveVel_B", "CurveVel_B_preprocessed")
```
